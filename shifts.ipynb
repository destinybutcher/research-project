```python
d = d.reset_index(drop=True)
d
employees = pd.DataFrame(data = {
    'id' : [0,1,2,3],
    'role' : ['offic','offic','barista','barista'],
    'offic_skills' : [100,60,10,20],
    'barista_skills' : [30,0,80,100],
})
```

```python
df = d.merge(
    employees,left_on='emp_id',right_on='id'
)
```

```python
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])
```

```python
df['working_units'] = (df['end'] - df['start']).dt.total_seconds() / 3600
```

```python
grouped = (
    df.groupby(['emp_id', 'role','offic_skills','barista_skills'], as_index=False)['working_units']
      .sum()
      .rename(columns={'working_units': 'total_working_units'})
)
```

```python
new_shifts_by_role = []
role = 'offic'
priority_by_role = grouped[grouped['role'] == role] # select only emp with particular role 
weekends = [('2025-10-18 07:00', '2025-10-18 15:00'), ('2025-10-19 07:00', '2025-10-18 12:00'), ('2025-10-18 11:00', '2025-10-18 15:00')]
sat = weekends[0]
sun = weekends[1:]
max_skills_by_role = priority_by_role[f'{role}_skills'].max() # find all emps with maximum skill for their role
sat_selected = priority_by_role[priority_by_role[f'{role}_skills'] == max_skills_by_role].sort_values(by='total_working_units') # select emp with least working hours

sat_shift = (sat,sat_selected.iloc[0]['emp_id'])
new_shifts_by_role.append(
    sat_shift
)
```

```python
first_half = priority_by_role[priority_by_role['emp_id'] != sat_shift[1]].sort_values(by=f'{role}_skills').iloc[0]['emp_id'] # работает не тот кто был в вс, но кто самый скиловые после него 
second_half = priority_by_role[priority_by_role['emp_id'] != first_half].sort_values(by='total_working_units').iloc[0]['emp_id'] # работает любой другой с наименьшими часами 
first_half,second_half
new_shifts_by_role.append(
    (sun[0],first_half)
)
new_shifts_by_role.append(
    (sun[0],second_half)
)
```

```python
# Run scheduler and emit logs/summary (role-specific windows)
import pandas as pd
from scheduler.config import load_config
from scheduler.data_io import read_employees, read_shifts, write_assignments
from scheduler.engine_baseline import greedy_schedule, build_requirements_for_day
from scheduler.validator import validate_assignments, summarize_assignments

print(f"[INFO] Loading config from {CONFIG_PATH}")
cfg = load_config(CONFIG_PATH)
print(f"[INFO] Busy days: {cfg.busy_days}")
print(f"[INFO] Loading employees from {EMPLOYEES_CSV}")
employees = read_employees(EMPLOYEES_CSV)
print(f"[INFO] Loading shifts from {SHIFTS_CSV} for week {WEEK_ID}")
shifts = read_shifts(SHIFTS_CSV, week_id=WEEK_ID)
print(f"[INFO] {len(shifts)} shifts found for week {WEEK_ID}")

print("[INFO] Generating assignments (greedy + backtracking, role-specific windows)")
assignments = greedy_schedule(employees, shifts, cfg)
print(f"[INFO] Generated {len(assignments)} assignments. Writing to {OUTPUT_CSV}")
write_assignments(OUTPUT_CSV, assignments)

# Build requirements by date for validation
req_by_date = {}
for date in sorted(shifts["date"].unique()):
    ds = pd.Timestamp(date).strftime("%Y-%m-%d")
    req_by_date[ds] = build_requirements_for_day(ds, cfg)

print("[INFO] Validating assignments...")
validate_assignments(
    employees,
    shifts,
    assignments,
    start_hm=cfg.default_shift.start,
    end_hm=cfg.default_shift.end,
    requirements_by_date=req_by_date,
)
print("[OK] Validation passed.")

print("\n===== SUMMARY =====")
print(summarize_assignments(assignments))

print("\nSample of assignments with shift/day types:")
print(assignments.head(10).to_string(index=False))
```

```python
import pandas as pd
from scheduler.config import load_config
from scheduler.engine_baseline import build_requirements_for_day

cfg = load_config(CONFIG_PATH)

# Employee availability by role
avail = employees['primary_role'].str.upper().value_counts()
print("[INFO] Available employees by role:\n", avail.to_string(), "\n")

# Requirements per date in the selected week
dates = sorted(shifts['date'].unique())
req_table = pd.DataFrame([build_requirements_for_day(pd.Timestamp(d).strftime("%Y-%m-%d"), cfg) for d in dates],
                         index=[pd.Timestamp(d).strftime("%Y-%m-%d") for d in dates]).fillna(0).astype(int)
print("[INFO] Requirements per day:\n", req_table, "\n")

# Quick feasibility check: if any role has 0 available but required > 0 on any day, that’s infeasible
impossible = []
for date_str, row in req_table.iterrows():
    for role, needed in row.items():
        have = int(avail.get(role, 0))
        if needed > 0 and have == 0:
            impossible.append((date_str, role, needed, have))
if impossible:
    print("[ERROR] Coverage impossible due to missing roles:")
    for date_str, role, need, have in impossible:
        print(f"  - {date_str}: {role} need={need}, have={have}")
else:
    print("[OK] No trivial impossibilities detected.")
```
